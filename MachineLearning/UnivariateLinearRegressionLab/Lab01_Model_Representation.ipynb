```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples: {m}")
```

```python
plt.scatter(x_train, y_train, marker='o', c ='r')
plt.title("house Prices")
plt.ylabel("Price ")
plt.xlabel('Size')
plt.show
```

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$

```python
w = 199.9929
b = 100.0116
print(f"w: {w}")
print(f"b: {b}")
```

```python
m = x_train.shape[0]
f = np.zeros(m)
f
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f = np.zeros(m)
    for i in range(m):
        f[i] = w * x[i] + b
    return f
```

```python
tmp_f = compute_model_output(x_train, w, b)
tmp_f
```

```python
plt.plot(x_train, tmp_f, c = 'b', label ='Prediction')
plt.scatter(x_train, y_train, marker='o', c ='r', label ='Actual Values')
plt.title("house Prices")
plt.ylabel("Price ")
plt.xlabel('Size')
plt.show
```
